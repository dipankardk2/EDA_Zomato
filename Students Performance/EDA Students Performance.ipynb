### Exploratory Data Analysis (EDA) of Students Performance

```python
# import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Avoid Warning message
import warnings
warnings.filterwarnings('ignore')
```

```python
#load dataset
data_real = pd.read_csv('Students Performance.csv')
```

```python
data_real.head()
```

```python
#shape of dataset
print(data_real.shape)
```

```python
#Columns
print(data_real.columns)
```

```python
print(data_real.info())
```

```python
#Check null values
print(data_real.isna().sum())
```

```python
#Rename column
data_real.rename(columns={'weight (in kgs)':'weight', 'height (in cms)':'height'}, inplace=True)
```

```python
#Check duplicate rows
print(data_real.duplicated().sum())
```

```python
#Check tutions column unique values
print(data_real['tutions'].unique())
```

```python
#replace 'none' by 'No tutions taken'
data_real['tutions'] = data_real['tutions'].replace('none', 'No Tutions Taken')
```

```python
print(data_real.tutions.unique())
```

```python
#Find the Male Female ratio in percentage 
value = data_real['gender'].value_counts()
label = data_real['gender'].unique()
```

```python
plt.figure(figsize=(6,6))
plt.pie(x=value, labels=label, autopct='%.2f%%', explode = (0.02,0.02),)
plt.show()
```

```python
#Section Column
y = data_real['section'].value_counts()
x = data_real['section'].unique()
sns.barplot(data=data_real, y=y, x=x, color='blue')
plt.xlabel('Section')
plt.ylabel('No of Students')
plt.title('Sections Wise Students Count', fontsize=12)
plt.show()
```

```python
#Check Parental Level Education
plt.figure(figsize=(8,8))
values = data_real['parental level of education'].value_counts(normalize=True)
label = data_real['parental level of education'].unique()
plt.pie(x=values, labels=label, autopct='%.2f%%', explode=(0.02,0.02,0.02,0.02,0.02,0.02))
plt.show()
```

```python
#Check weight column using violinplot
sns.violinplot(x = 'weight (in kgs)', data=data_real)
```

```python
#Check weight column using binning
value = data_real['weight'].value_counts(bins=4)
print(value)
#Maximum students weight is from 55 kgsto 65 kgs
```

```python
sns.violinplot(y='weight', data=data_real, x='gender')
```

```python
#Box Plot for various subjects
plt.figure(figsize=(12,5))
data_real.boxplot(column=['math score','english score','science score', 'social studies score',
       'computer application score'])
```

```python
#Categorical Plot
for col in data_real.columns[6:11]:
    sns.catplot(data=data_real, x='house', y=col)
```

```python
# Visualization of numerical columns
for col in data_real.columns[6:12]:
    print('Dist Plot for '+col)
    sns.distplot(data_real[col], kde=True, bins=10)
    plt.show()
```

```python
# Swarm plot for Subject score
for col in data_real.columns[6:11]:
    print('Swarm Plot for '+col)
    sns.swarmplot(x='house', y=col, hue='gender', data=data_real, dodge=True)
    plt.show()
```

```python
# Point plot for Subject marks
for col in data_real.columns[6:11]:
    print('Point Plot for '+col)
    sns.pointplot(x='section', y=col, hue='gender', dodge=True, data=data_real)
    plt.show()
```

```python
#Check Correlation between features
plt.figure(figsize=(8,5))
sns.heatmap(data_real.corr(), fmt='0.2g', annot=True)
```

```python
#Regression Plot
sns.regplot(x='math score', y='science score', data=data_real,marker= '*', color='g' )
```

```python
#Regression Plot
sns.regplot(x='english score', y='science score', data=data_real,marker= '*', color='b' )
```

```python
#Regression Plot
sns.regplot(x='english score', y='math score', data=data_real,marker= 'o', color='m' )
```
